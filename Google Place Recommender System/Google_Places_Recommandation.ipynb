```python
#Preprocessing:

reviews = reviews[['rating', 'reviewerName', 'reviewText', 'categories', 'jobs', 'currentPlace', 'previousPlaces','address', 'education', 'name', 'price', 'closed']]

reviews.shape
```

```python
reviews.isnull().sum()
```

```python
reviews['reviewerName'].fillna('User', inplace = True)

reviews.shape
```

```python
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
      l.append(i)
  return l
```

```python
reviews['reviewText'] = reviews['reviewText'].apply(lambda x:x.split())

reviews.head()
```

```python
reviews['categories'] = reviews['categories'].apply(convert)

reviews.head()
```

```python
reviews.shape
```

```python
reviews['jobs'].fillna('No Job', inplace = True)

reviews['jobs'] = reviews['jobs'].apply(convert)

reviews['jobs'] = reviews['jobs'].apply(lambda x: ' '.join(map(str, x)))

reviews.head()
```

```python
reviews['currentPlace'] = reviews['currentPlace'].apply(convert)

reviews['currentPlace'] = reviews['currentPlace'].apply(lambda x: ' '.join(map(str, x)))

reviews.head()
```

```python
reviews['previousPlaces'] = reviews['previousPlaces'].apply(convert)

reviews['previousPlaces'] = reviews['previousPlaces'].apply(lambda x: ' '.join(map(str, x)))

reviews.head()
```

```python
reviews['address'] = reviews['address'].apply(convert)

reviews['address'] = reviews['address'].apply(lambda x: ' '.join(map(str, x)))

reviews.head()
```

```python
reviews['education'] = reviews['education'].apply(convert)

reviews['education'] = reviews['education'].apply(lambda x: ' '.join(map(str, x)))

reviews.head(1)
```

```python
#Label Encoding:
le = preprocessing.LabelEncoder()
reviews['closed']= le.fit_transform(reviews['closed'])
reviews['closed'].unique

reviews.head()
```

```python
reviews['categories'] = reviews['categories'].apply(lambda x: [i.replace(" ", "") for i in x])

reviews.head()
```

```python
reviews['tags'] = reviews['reviewText'].astype(str) + reviews['jobs'].astype(str) + reviews['education'].astype(str)
```

```python
reviews['place'] = reviews['currentPlace'].astype(str) + reviews['previousPlaces'].astype(str) + reviews['address'].astype(str)
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df['place'] = df['place'].apply(lambda x:x.lower())
```

```python
ps = PorterStemmer()

def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python
df['place'] = df['place'].apply(stem)
```

```python
df.head(10)
```

```python
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df['tags'].values[0:10000], df['categories'].values[0:10000]).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
similarity = cosine_similarity(vectors)
similarity
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key= lambda x: x[1])[1:6]
```

```python
def recommand(names):
  reviews_index = df[df['name'] == names].index[0]
  distances = similarity[reviews_index]
  reviews_list = sorted(list(enumerate(distances)), reverse = True, key= lambda x: x[1])[1:6]
#   print(reviews_list)  
  for i in reviews_list:
#      print(df.iloc[i[1]].name)
    print(df['name'][i[0]]),print(i)
```
